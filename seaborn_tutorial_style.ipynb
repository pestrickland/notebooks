## Aesthetics

### Basics

```python
np.random.seed(sum(list(map(ord, "aesthetics"))))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
import seaborn as sns
sinplot()
```

### Styling

There are five preset themes:

* `darkgrid`
* `whitegrid`
* `dark`
* `white`
* `ticks`

The default is `darkgrid`.

```python
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data);  # Semicolon supresses output of figure object.
```

```python
sns.set_style("dark")
sinplot()
```

```python
sns.set_style("white")
sinplot()
```

```python
sns.set_style("ticks")
sinplot()
```

### Removing and offsetting spines

```python
sinplot()
sns.despine()
```

```python
f, ax = plt.subplots()
sns.offset_spines()
sns.violinplot(data)
sns.despine(trim=True)
```

`trim` removes the excess range from the spines (note how the x-axis starts at 1 and ends at 6)

```python
sns.set_style("whitegrid")
sns.boxplot(data, color="deep")
sns.despine(left=True)
```

### Temporarily setting figure style

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

### Creating custom palettes

A single colour can be specified to generate a palette that blends the colour to dark grey.

```python
sns.palplot(sns.dark_palette("MediumPurple"))
```

```python
sns.palplot(sns.dark_palette("#e74c3c"))
```

```python
sns.palplot(sns.dark_palette("skyblue", 8, reverse=True))
```

By default, the list of colours is returned, but it can also be returned as a colormap object to be passed to plotting functions.

```python
sample = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=100)
pal = sns.dark_palette("MediumPurple", as_cmap=True)
plt.figure(figsize=(6, 6))
sns.kdeplot(sample, cmap=pal);
```

By appending "_d" to the name of a sequential ColorBrewer palette, a darker palette is created that remains harmonious to the base palette but is better suited to drawing line or contour plots.

```python
sns.palplot(sns.color_palette("BuPu"))
```

```python
sns.palplot(sns.color_palette("BuPu_d"))
```

The `blend_palette()` function is a more general way to produce palettes. It interpolates between a number of seed points.

```python
sns.palplot(sns.blend_palette(["mediumseagreen", "ghostwhite", "#4168B7"], 9))
```

A sequential palette can be created along a saturation scale:

```python
sns.palplot(sns.blend_palette([sns.desaturate("#009B76", 0), "#009B76"], 5))
```

An ordered palette that can colour large patches:

```python
sns.palplot(sns.blend_palette(["seagreen", "lightblue"]))
```

### Using palettes in seaborn functions

```python
gammas = sns.load_dataset("gammas")
sns.set_style("darkgrid")
sns.tsplot(gammas, "timepoint", "subject", "ROI", "BOLD signal", color="muted")
```

```python
gammas = sns.load_dataset("gammas")
sns.tsplot(gammas, "timepoint", "subject", "ROI", "BOLD signal", color="deep")
```

```python
tips = sns.load_dataset("tips")
pal = sns.blend_palette(["seagreen", "lightblue"], 4)
sns.violinplot(tips.total_bill, tips.day, color=pal)
```

```python
sns.factorplot("day", "total_bill", hue="smoker", data=tips, kind="box", palette="OrRd");
```

```python
attend = sns.load_dataset("attention")
sns.factorplot("solutions", "score",
               hue="subject", col="attention",
               data=attend, ci=None,
               legend=False, palette="Paired")
```
