# Visualising distributions

## Histograms

```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
```

```python
sns.set_palette("deep", desat=0.6)
sns.set_context(rc={"figure.figsize": (8, 4)})
np.random.seed(9221999)
```

```python
data = randn(75)
plt.hist(data);
```

The default behaviour is for 10 bins of equal widths.

```python
plt.hist(data, 6, color=sns.desaturate("indianred", 0.75));
```

Using the `normed` argument, distributions with an unequal number of observations can be compared.

```python
data1 = stats.poisson(2).rvs(100)
data2 = stats.poisson(5).rvs(120)
max_data = np.r_[data1, data2].max()
bins = np.linspace(0, max_data, max_data + 1)
plt.hist(data1, bins, normed=True, color="#6495ed", alpha=0.5)
plt.hist(data2, bins, normed=True, color="#f08080", alpha=0.5);
```

```python
x = stats.gamma(3).rvs(5000)
plt.hist(x, 70, histtype="stepfilled", alpha=0.7);
```

A joint distribution can be plotted using `jointplot`.

```python
y = stats.gamma(5).rvs(5000)
sns.jointplot(x, y, kind="hex");
```

## Estimating density of observations

A rug plot simply draws a vertical line at each observed data point.

```python
sns.set_palette("hls", 1)
data = randn(30)
sns.rugplot(data)
plt.ylim(0, 1);
```

If we tie the rug plot to the histogram, we can see how representative the histogram is.

```python
plt.hist(data, alpha=0.3)
sns.rugplot(data);
```

A kernel density plot is also a transformation from simple tick-marks to a height-encoded measure of density. But instead of simply binning each mark, each will be represented with a gaussian basis function.

```python
# Draw the rug and set up the x-axis space.
sns.rugplot(data);
xx = np.linspace(-4, 4, 100)

# Compute the bandwidth of the kernel using a rule-of-thumb.
bandwidth = ((4 * data.std() ** 5) / (3 * len(data))) ** .2
bandwidth = len(data) ** (-1. / 5)

# We'll save the basis functions for the next step.
kernels = []

# Plot each basis function.
for d in data:
    # Make the basis function as a gaussian PDF.
    kernel = stats.norm(d, bandwidth).pdf(xx)
    kernels.append(kernel)

    # Scale for plotting.
    kernel /= kernel.max()
    kernel *= .4
    plt.plot(xx, kernel, "#888888", alpha=.5)
plt.ylim(0, 1);
```

The basis functions are then summed and plotted. SciPy can also perform a kernel density estimate, giving similar results as shown below.

```python
# Set up the plots.
f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
c1, c2 = sns.color_palette("husl", 3)[:2]

# Plot the summed basis functions.
summed_kde = np.sum(kernels, axis=0)
ax1.plot(xx, summed_kde, c=c1)
sns.rugplot(data, c=c1, ax=ax1)
ax1.set_yticks([])
ax1.set_title("Summed Basis Functions")

# Use scipy to get the density estimate.
scipy_kde = stats.gaussian_kde(data)(xx)
ax2.plot(xx, scipy_kde, c=c2)
sns.rugplot(data, c=c2, ax=ax2)
ax2.set_yticks([])
ax2.set_title("Scipy gaussian_kde")
f.tight_layout()
```

Seaborn has a high-level function to plot a kernel density estimate in one step:

```python
sns.kdeplot(data, shade=True);
```

The bandwidth of the KDE can be adjusted, as shown below.

```python
pal = sns.blend_palette([sns.desaturate("royalblue", 0), "royalblue"], 5)
bws = [0.1, 0.25, 0.5, 1, 2]

for bw, c in zip(bws, pal):
    sns.kdeplot(data, bw=bw, color=c, lw=1.8, label=bw)
    
plt.legend(title="Kernel Bandwidth")
sns.rugplot(data, color="#333333");
```

Gaussian kernels are the most common and probably most useful, but there are a variety of other kernels that can be used. Other kernels require the `statsmodel` package to be installed, which in turn requires `patsy`, although it doesn't get installed as a dependency for some reason.

```python
kernels = ["biw", "cos", "epa", "gau", "tri", "triw"]
pal = sns.color_palette("hls", len(kernels))

for k, c in zip(kernels, pal):
    sns.kdeplot(data, kernel=k, color=c, label=k)
plt.legend();
```
