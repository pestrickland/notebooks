`prettyplotlib` is a library designed to produce pretty matplotlib plots. It is described [here](http://blog.olgabotvinnik.com/post/58941062205/prettyplotlib-painlessly-create-beautiful-matplotlib) and installed using pip.

By default, the following type of plot is produced.

```python
# Set a random seed for consistent results later on.
np.random.seed(15)

fig, ax = plt.subplots(1)

# Show the whole colour range.
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    ax.scatter(x, y, label=str(i))
ax.legend()
ax.set_title("Scatter example showing \ndefault matplotlib scatter")
```

Without using the `prettyplotlib` library, it is still possible to produce better-looking plots. However, an additional module is used, `brewer2mpl`, that gets high quality colour schemes as described at <http://colorbrewer2.org>.

```python
import brewer2mpl

# Get "Set2" colours from ColorBrewer (all ColorBrewer scales are available
# at <http://bl.ocks.org/mbostock/5577023>)
set2 = brewer2mpl.get_map("Set2", "qualitative", 8).mpl_colors

# Set the random seed again for consistency.
np.random.seed(15)

fig, ax = subplots(1)

# Show the whole colour range.
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    colour = set2[i]
    ax.scatter(x, y, label=str(i), color=colour)
ax.legend()
ax.set_title("Scatter example showing \n matplotlib scatter \nwith "
             "ColorBrewer colours")
```

It is clear now that there are eight separate groups of data on this chart, and they are quite distinguishable from one another. If we add some transparency we get the following:

```python
fig, ax = subplots(1)

# Set the random seed again for consistency.
np.random.seed(15)

# Show the whole colour range.
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    colour = set2[i]
    ax.scatter(x, y, label=str(i), alpha=0.5, color=colour)
ax.set_title("Scatter example showing \n matplotlib scatter \nwith "
             "ColorBrewer colours and alpha=0.5")
```

Sometimes we don't want to have the top and right hand axes, so we could remove them:

```python
fig, ax = subplots(1)

# Set the random seed again for consistency.
np.random.seed(15)

# Show the whole colour range.
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    colour = set2[i]
    ax.scatter(x, y, label=str(i), alpha=0.5, facecolor=colour, 
               edgecolor="black", linewidth=0.15)
ax.set_title("Scatter example showing \n matplotlib scatter \nwith "
             "ColorBrewer colours and alpha=0.5")

# Remove top and right axes lines.
spines_to_remove = ["top", "right"]
for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
# Remove ticks from top and right axes lines.
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
```

We can do more; let's thin the axis lines and make them, and the edges of the scatter markers, less black:

```python
fig, ax = subplots(1)

# Set the random seed again for consistency.
np.random.seed(15)

# Choose a dark grey to replace black.
almost_black = "#262626"

# Show the whole colour range.
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    colour = set2[i]
    ax.scatter(x, y, label=str(i), alpha=0.5, facecolor=colour, 
               edgecolor=almost_black, linewidth=0.15)
ax.set_title("Scatter example showing \n matplotlib scatter \nwith "
             "ColorBrewer colours and alpha=0.5")

# Remove top and right axes lines.
spines_to_remove = ["top", "right"]
for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
# Remove ticks from top and right axes lines.
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")

# Thin remaining spines and change colour to dark grey.
spines_to_keep = ["left", "bottom"]
for spine in spines_to_keep:
    ax.spines[spine].set_linewidth(0.5)
    ax.spines[spine].set_color(almost_black)
```

Some improvements:

* Remove the border
* Change the text colour
* Change three points to one for each group of data
* Fill the legend with a light grey background

```python
fig, ax = subplots(1)

# Set the random seed again for consistency.
np.random.seed(15)

# Choose a dark grey to replace black.
almost_black = "#262626"

# Show the whole colour range.
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    colour = set2[i]
    ax.scatter(x, y, label=str(i), alpha=0.5, facecolor=colour, 
               edgecolor=almost_black, linewidth=0.15)
ax.set_title("Scatter example showing \n matplotlib scatter \nwith "
             "ColorBrewer colours and alpha=0.5")

# Remove top and right axes lines.
spines_to_remove = ["top", "right"]
for spine in spines_to_remove:
    ax.spines[spine].set_visible(False)
    
# Remove ticks from top and right axes lines.
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")

# Thin remaining spines and change colour to dark grey.
spines_to_keep = ["left", "bottom"]
for spine in spines_to_keep:
    ax.spines[spine].set_linewidth(0.5)
    ax.spines[spine].set_color(almost_black)
    
# Change tick labels to almost_black.
ax.xaxis.label.set_color(almost_black)
ax.yaxis.label.set_color(almost_black)

# Change title to almost_black.
ax.title.set_color(almost_black)

# Remove the border of the legend box and fill it with light grey.
light_grey = np.array([240/255] * 3)
legend = ax.legend(frameon=True, scatterpoints=1)
rect = legend.get_frame()
rect.set_facecolor(light_grey)
rect.set_linewidth(0.0)

# Change the legend labels to almost_black.
texts = legend.texts
for t in texts:
    t.set_color(almost_black)
    
# Alpha to the legend can also be added.
rect.set_alpha(0.8)
```

The `prettyplotlib` library can accomplish all of that code in a quicker way, although you should note that the *all* inner ticks have been removed.

```python
import prettyplotlib as ppl

# This is "import matplotlib.pyplot as plt" from the prettyplotlib library
from prettyplotlib import plt

# This is "import matplotlib as mpl" from the prettyplotlib library
from prettyplotlib import mpl

# Set the random seed for consistency.
np.random.seed(15)

fig, ax = plt.subplots(1)

# Show the whole colour range.
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    ppl.scatter(ax, x, y, label=str(i))
    
ppl.legend(ax)
ax.set_title("prettyplotlib scatter example \nshowing default colour cycle "
             "\nand scatter parameters")
```
